```python
import sympy
from sympy import *
```

```python
init_printing(use_unicode=False, wrap_line=False)
x = Symbol('x')
u = Symbol('u')
v = Symbol('v')
w3 = Symbol('w3')
w2 = Symbol('w2')
w1 = Symbol('w1')
w0 = Symbol('w0')
```

```python
# relu_x = Piecewise((0,x<0),(x, x>=0))
# relu_x
```

```python
norm = exp(-(x - u)**2/2/v**2) /v/sqrt(2 * pi)
```

```python
apprx_3rd = w3 * x**3 + w2 * x**2 + w1 * x + w0
```

```python
loss = sympy.simplify(integrate(norm * x**2, (x, 0, +oo))) + sympy.simplify(integrate(norm * apprx_3rd**2, (x, -oo, +oo))) - \
    sympy.simplify(integrate(2 * norm * x*apprx_3rd, (x, 0, +oo)))
```

```python
loss_sim = sympy.simplify(loss)
```
